```python
import math
import matplotlib.pyplot as plt
import numpy as np
import timeit
```

```python
x = np.array(range(-10,10))
y = x**4 - 10
plt.plot(x,y)
plt.grid()
plt.show()
```

```python

def f(x):
    return x**4 - 10
a = 1
b = 2
tolerance = 1 *(10**-6)
max_iteration = 100
```

```python
for i in range(max_iteration):
    c = (a+b)/2
    print(f"Iteration number: {i}")
    if abs(f(c)) < tolerance:
        print(f"Root found at x = {c:.6f}")  
        print("Time taken: ",timeit.timeit())
        break
    elif f(c)*f(a)<0:
        b=c
    else:
        a=c
```
